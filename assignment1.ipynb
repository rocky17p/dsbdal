```python
df['Car']=df['Car'].fillna(df['Car'].mode)
df['BuildingArea']=df['BuildingArea'].fillna(df['BuildingArea'].mean)
df['YearBuilt']=df['YearBuilt'].fillna(df['YearBuilt'].mode)
df['CouncilArea']=df['CouncilArea'].fillna(df['CouncilArea'].mode)
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['Address']=df['Address'].astype('string')
df['Suburb']=df['Suburb'].astype('string')
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
normalized_price=StandardScaler().fit_transform(df[['Price']])
```
